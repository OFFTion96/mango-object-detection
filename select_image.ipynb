```python
def get_img_except(l1,l2):
    img_exept = []
    for i in l1:
        if i not in l2:
            img_exept.append(i)
    return img_exept
            
```

```python
img_tata_except = get_img_except(img_tata,inter_img_tata)
img_aj_except = get_img_except(img_aj,inter_img_aj)
img_mutu_except = get_img_except(img_mutu,inter_img_mutu)
```

```python
def get_img_to_annotation(list_exept,list_raw):
    img_path=[]
    for i in list_raw:
        img_raw = re.search("IMG_(\d+)",i)
        if (img_raw!= None):
            if img_raw[0] in list_exept:
                img_path.append(i)
    return img_path
```

```python
new_img_tata = get_img_to_annotation(img_tata_except,path_img_tata)
```

```python
new_img_aj = get_img_to_annotation(img_aj_except,path_img_aj)
```

```python
new_img_mutu = get_img_to_annotation(img_mutu_except,path_img_mutu)
```
